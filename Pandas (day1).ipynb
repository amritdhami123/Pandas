<h2> <b> Filtering </b> </h2>

```python
data
```

```python
data.columns
```

```python
# filter all the data whose unit price > 200 
con = data['UNIT PRICE '] > 300
```

```python
data[con]
```

```python
# multiple condition 
# slect unit price > 300 and items sold >500 
con1= data['UNIT PRICE ']>300
con2 = data['ITEMS SOLD ']>500
```

```python
data[con1 & con2] 
```

```python
# OR Condition 
# slect unit price > 300 or items sold >500 
data[con1 | con2] # or '|' (pipe sign) 
```

```python
# Not Conditon 
# select all the data where state is not equal to uttar prades 
```

```python
# in 
# select all the values from city 'Pune' and 'Simla' 
con =['Simla', 'Pune']
# data['city'] ='Pune' | data['city'] ='Simla' ] 
df[df['City'].isin(con)]
```
