```python
# mean() 
mean = data['power'].mean()
mean
```

```python
data['power'] = data['power'].fillna(mean)
data['power'].isna().sum()
```

```python
data.isna().sum()
```

```python
# fill the 'na' value in engine with 'backwardfill' 
data['engine'] = data['engine'].fillna(method='bfill')
```

```python
# maualvalue 
```

```python
data['seats']= data['seats'].fillna(5)
```

```python
data['mileage'] = data.mileage.fillna(method='ffill')
```
