<h1> <b> Pivot Table </b> </h1> 

```python
data.head(2)
```

```python
# pivot_table
```

```python
pivoted = pd.pivot_table(data, 
              values='mileage', 
              index='location',
              aggfunc=['min','max','mean'])
```

```python
pivoted
```

```python
# Grouby 
df_x = data.groupby('location')['mileage'].agg(['min', 'max', 'mean'])
```

```python
df_x.to_csv('grouped_data.csv')
```

```python
# Unpivot Data 
```

```python
pd.melt(pivoted, id_vars = ['location'], value_vars=['mileage'], var_name='location', value_name='location')
```
