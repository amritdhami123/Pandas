```python
data = sns.load_dataset('iris')
data.species.unique()
```

```python
data
```

```python
# X - all input 
# y output 
X=data.drop('species', axis=1) 
X.head()
```

```python
y = data['species'] 
y.head(2)
```

```python
#data 100% 
# 80% train 
# 20 % test 
```

```python
# train_test_split 
from sklearn.model_selection import train_test_split
```

```python
X X_trian , X_test
y y_train, y_test 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Model Train 
#import algorithm 
from sklearn.svm import SVC # it import support v m classifier 
```

```python
# Model 
model = SVC()
```

```python
model
```

```python
# Model Train 
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred
```

```python
y_test
```

```python
# Checking Accuracy 
from sklearn.metrics import accuracy_score
```

```python
acs = accuracy_score(y_pred, y_test)
acs
```
